#### Armazenar os balanços e DREs em um dicionário onde a chave é o nome da empresa

```python
import pandas as pd
import os ##para utilizar arquivos da máquina

# fundamentos = {
#     "ABEV3": balanco_dre_abev3,
#     "MGLU3": balanco_dre_mglu3
# }
dicionario_de_fundamentos = {}
arquivos = os.listdir("balancos")

contador_de_empresas = 0

for arquivo in arquivos:
    #percorrer todos os arquivos de balanço e tirar o nome "balanço" e ".xls" do nome para facilitar o trabalho
    nome = arquivo
    limpar_nome_arquivo = "balanco_.xls"
    
    for i in limpar_nome_arquivo:
        nome = nome.replace(i,'')
    
    if nome in lista_de_empresas:
        #print(nome)
        contador_de_empresas = contador_de_empresas + 1
        #pegar o balanço das empresas listadas
        balanco = pd.read_excel(f'balancos/{arquivo}', sheet_name=0)
        # na primeira coluna colocar o título com o nome da empresa
        balanco.iloc[0, 0] = nome
        # pegar 1ª linha e tornar um cabeçalho
        balanco.columns = balanco.iloc[0]
        balanco = balanco[1:]
        # tornar a 1ª coluna (que agora tem o nome da emrpesa)
        balanco = balanco.set_index(nome)
        
        dicionario_de_fundamentos[nome] = balanco
        
        
print("TOTAL DE EMPRESAS: ", contador_de_empresas)
```

#### Eliminando empresas que o último balanço divulgado era anterior a 01/01/2021

#### Armazenar cotações em um dicionário onde a chave é o nome da empresa

```python
import pandas as pd
import os ##para utilizar arquivos da máquina

dicionario_de_cotacoes = {}
arquivos = os.listdir("cotacoes")

contador_de_empresas = 0
empresas_para_excluir = []

for arquivo in arquivos:
    #percorrer todos os arquivos de cotações e tirar o nome "cotacoes" e ".csv" do nome para facilitar o trabalho
    nome = arquivo
    limpar_nome_arquivo = "cotacoes_.csv"

    for i in limpar_nome_arquivo:
        nome = nome.replace(i,'')
    
    if nome in lista_de_empresas:
        #print(nome)
        contador_de_empresas = contador_de_empresas + 1
        cotacao = pd.read_csv(f'cotacoes/{arquivo}')
        dicionario_de_cotacoes[nome] = cotacao
        #print("EMPRESA: ",nome,"\n")
    
    else:
        #Se foi feito download de uma cotação de uma empresa que já foi excluída da análise ou muito recente na bolsa, ela
        #será excluída
        empresas_para_excluir.append(f'{nome}')
        
        
print("\nNÚMERO DE EMPRESAS: ", contador_de_empresas)
print("\nNÚMERO DE EMPRESAS EXCLUÍDAS: ",len(empresas_para_excluir))

if len(empresas_para_excluir) > 0:
    print("\nEMPRESAS EXCLUÍDAS: \n", empresas_para_excluir)
```

#### Definir o período dos dados (janeiro de 2016 - julho de 2021) para que os dados trabalhados pertençam a esse período
Ou seja, eliminar empresas que ainda estão presentes e não possuem dados neste período, além de eliminar dados referentes a períodos mais antigos (anteriores a janeiro de 2016) de empresas que possuírem dados no período a ser analisado.

```python
data_inicio = "01/01/2016"
lista_para_apagar_fundamentos = []

data_aux = pd.to_datetime(data_inicio)

for empresa in lista_de_empresas_atualizada:
    lista_para_apagar_fundamentos.clear()
    colunas = list(dicionario_de_fundamentos[empresa].columns)
    
    for coluna in colunas:
        colunaAux = pd.to_datetime(coluna)
        
        if colunaAux < data_aux:
            lista_para_apagar_fundamentos.append(coluna)
    
    for auxiliar in lista_para_apagar_fundamentos:
        dicionario_de_fundamentos[empresa].pop(auxiliar)
    
    print("EMPRESA: ", empresa)
    print("\nCOLUNAS PARA APAGAR: ", lista_para_apagar_fundamentos,"\n")
```

#### Alterar os valores para percentual
Fundamento Percentual do Trimestre = Fundamento Trimestre / Fundamento Trimestre Anterior

Cotação Percentual do Trimestre = Cotação Trimestre Seguinte / Cotação Trimestre

A classificação de um trimestre para um papel pode ser de compra ("Buy", com valor 2), de segurar a ação ("Hold", com valor 1) ou de venda ("Sell", com valor 0)

- BUY -> Se a variação da cotação no trimestre seguinte superar o índice IBOV em 3% ou mais;
- SELL -> Se a variação da cotação no trimestre for menor do que o índice IBOV em 3% ou menos;
- HOLD -> Se nenhuma das duas situações acontecer.

```python
for empresa in dicionario_final:
    tabela = dicionario_final[empresa]
    tabela.drop(index=tabela.index[0], axis=0, inplace=True)
    for coluna in tabela:
        if "Adj Close" in coluna or "IBOV" in coluna or "Media Cotacoes Setor" in coluna:
            pass
        else:
            #pegar o fundamento anterior
            condicoes = [
                (tabela[coluna].shift(1) > 0) & (tabela[coluna] < 0),
                (tabela[coluna].shift(1) < 0) & (tabela[coluna] > 0),
                (tabela[coluna].shift(1) < 0) & (tabela[coluna] < 0),
                (tabela[coluna].shift(1) == 0) & (tabela[coluna] > 0),
                (tabela[coluna].shift(1) == 0) & (tabela[coluna] < 0),
                (tabela[coluna].shift(1) < 0) & (tabela[coluna] == 0),
            ]
            valores = [
                -1,
                1,
                (abs(tabela[coluna].shift(1)) - abs(tabela[coluna])) / abs(tabela[coluna].shift(1)),
                1,
                -1,
                1,
            ]
            
            tabela[coluna] = np.select(condicoes, valores, default=tabela[coluna] / tabela[coluna].shift(1) - 1)
    
    #pegar a cotação seguinte
    tabela["Adj Close"] = tabela["Adj Close"].shift(-1) / tabela["Adj Close"] - 1
    tabela["IBOV"] = tabela["IBOV"].shift(-1) / tabela["IBOV"] - 1
    tabela["Resultado"] = tabela["Adj Close"] - tabela["Media Cotacoes Setor"]
    condicoes = [
        (tabela["Resultado"] > 0.03),
        (tabela["Resultado"] < 0.03) & (tabela["Resultado"] > -0.03),
        (tabela["Resultado"] < -0.03)
    ]
    valores = [2, 1, 0] ########ALTERADO DE [2,1,0] PARA [1,2,0] PARA TESTAR AS REDES NEURAIS 
    tabela["Decisao"] = np.select(condicoes, valores)
            
    dicionario_final[empresa] = tabela

display(dicionario_final["ABEV3"])
```
